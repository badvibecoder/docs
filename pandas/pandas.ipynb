```python
# Apply a filter using columns
df[df["Category"] == "Electronics"]

# df[] in my dataframe, I want to look for all values in 
# dataframe column Category, that equal Electronics
```

```python
# Apply a filter using columns with multiple filters
df[(df["Category"] == "Electronics") & (df["Country"] == "USA")]
```

```python
# Apply a filter using columns with multiple filters, OR
df[(df["Category"] == "Electronics") | (df["Country"] == "USA")]
```

```python
df[df["Quantity"] >= 5]
```

```python
# Locate a customer name that starts with Z

df[df["CustomerName"].str.startswith("A")]
```

```python
# Locate a customer name that starts with Z

df[df["CustomerName"].str.endswith("a")]
```

```python
# Look for data in a setup of elements
df[df["Country"].isin(["USA", "Sweden", "Japan"])]
```

```python
# Use the ~ to reverse any condition in pandas
df[~df["Country"].isin(["USA", "Sweden", "Japan"])]
```

```python
# 21:43
# https://www.youtube.com/watch?v=EXIgjIBu4EU

# Update Data

df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] = "Laser Clock"

# Use Locate to filter the column and the data, then comma, 
# and the data you want to edit in that row
```

```python
df.loc[df["CustomerName"] == "Anna Ivanova"]
```

```python
# We can make bulk changes as well, change all USA to United States

df.loc[df["Country"] == "USA", "Country"] = "United States"
```

```python
df[df["Country"].isin(["United States"])]
```

```python
df["Country"] = df["Country"].str.upper()
```

```python
df["Country"]
```

```python
# Remove index position 39
df = df.drop(39)
df["Country"]
```

```python
# drop anything with null values
df.dropna()

# fill missing / null with something
df.fillna({"OrderID": 0}, inplace=True)
```

```python
# Rename Columns
df.rename(columns={"OrderID": "Order ID"}, inplace=True)
```
