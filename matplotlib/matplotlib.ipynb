### Matplotlib Architecture

- Figure
    - top level container, the window essentially/canvas
- Axes
    - The plot data that is drawn on, multiple axes can exist on a figure
- Axis
    - The number line object, x and y axis


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# automatically graph inline
%matplotlib inline

# set style if needed
plt.style.use('dark_background')

# Method 1 pyplot apprach
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')
plt.show() # Not needed in jupyterlab
```

```python
plt.plot(x,y)
plt.plot(y,x) # Added another line
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')
plt.show() # Not needed in jupyterlab
```

```python
# Use subplots for more control

fig, ax = plt.subplots() # Creates a figure and axes object
ax.plot(x,y)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Line Plot (Object Oriented)')
plt.show()

# ax is now its only painting you can add to the canvas

```

```python
# Use subplots for more control

fig, ax = plt.subplots() # Creates a figure and axes object
fig, ay = plt.subplots() # Creates a figure and axes object
ax.plot(x,y)
ay.plot(y,x)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ay.set_xlabel('X Axis')
ay.set_ylabel('Y Axis')
ax.set_title('Line Plot X (Object Oriented)')
ay.set_title('Line Plot Y (Object Oriented)')
plt.show()

# ax is now its only painting you can add to the canvas
```

```python
# Basic line plots
# 8:45

# Sample data
x = np.linspace(0, 10, 100) # creates 100 points from 0 to 10
y = np.sin(x) # y is the sine wave of x

plt.figure(figsize=(8,5)) # width x height
plt.plot(x,y)
plt.xlabel('X Values')
plt.ylabel('sin(X)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
# multiple lines on the same plot

# Sample data
x = np.linspace(0, 10, 100) # creates 100 points from 0 to 10
y1 = np.sin(x) # y is the sine wave of x
y2 = np.cos(x) # y is the cosign of x
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(8,5)) # width x height
plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='cos(x)')
plt.plot(x,y3, label='sin(x) * cos(x)')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Multiple Functions')
plt.legend() # Show the legend
plt.grid(True, alpha=0.3) # Set transparency
plt.show()
```

### Customizing Styles

```python
x = np.linspace(0,10,20)

plt.figure(figsize=(12,6))

plt.plot(x,x, 'b-', label='Solid Line', linewidth=2)
plt.plot(x,x+1, 'r--', label='Dashed Line', linewidth=2)
plt.plot(x,x+2, 'g-', label='Dash-dot Line', linewidth=2)
plt.plot(x,x+3, 'm:', label='Dotted Line', linewidth=2)
plt.plot(x,x+4, 'co-', label='Circles', markersize=8)
plt.plot(x,x+5, 's-', label='Squares', markersize=6)
plt.plot(x,x+6, '^-', label='Triangles', markersize=6)

# you can specify with color='', linestyle='' etc.. as well
plt.plot(x,x+7, linestyle='dashdot', color='pink', linewidth=6, label='custom' )
#same as linestyle='-.'

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Styles and Markers')
plt.legend() # Show the legend
plt.grid(True, alpha=0.3) # Set transparency
plt.show()
```
